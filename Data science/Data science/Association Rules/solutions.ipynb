#  Import necessary libraries

```python
import pandas as pd
```

#  Load the dataset


```python
file_path = 'Online retail.xlsx'
df = pd.read_excel(file_path)
```

#  Show first few rows to understand the data


```python
df.head()
```

```python
# Split each row into a list of items
transactions = df.iloc[:, 0].apply(lambda x: x.split(','))

# Convert to list of lists
transactions_list = transactions.tolist()

# Get a list of all unique items
all_items = set(item.strip() for sublist in transactions_list for item in sublist)

# ðŸ”§ Convert set to list
all_items = list(all_items)

# Create a DataFrame with one column per item
encoded_df = pd.DataFrame(0, index=range(len(transactions_list)), columns=all_items)

# Fill with 1 where item is present
for i, items in enumerate(transactions_list):
    for item in items:
        encoded_df.at[i, item.strip()] = 1

# Show first few rows
encoded_df.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# ðŸ”Ž Find frequent itemsets with min_support threshold
frequent_itemsets = apriori(encoded_df, min_support=0.02, use_colnames=True)

# ðŸ’¬ Generate the rules from frequent itemsets
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Show the rules
rules.head()
```
